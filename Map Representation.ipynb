```python
import pandas as pd
```

```python
#Portal de datos abiertos de Madrid - Censo de locales
localesdf = pd.read_csv("/home/seb/Downloads/Locales201606.csv",sep=";")
```

```python
localesdf.head(5)
```

```python
len(localesdf)
```

```python
localesdf['coordenada_x_local']=pd.to_numeric(localesdf['coordenada_x_local'].str.replace(",","."))
localesdf['coordenada_y_local']=pd.to_numeric(localesdf['coordenada_y_local'].str.replace(",","."))
```

```python
localesdf=localesdf[localesdf['coordenada_x_local']>100000]
```

```python
localesdf.head(32)
```

```python
import utm
localesdf['latitude']=localesdf.apply(lambda row: utm.to_latlon(row['coordenada_x_local'],row['coordenada_y_local'],30,"T")[0] , 1)   
localesdf["longitude"]=localesdf.apply(lambda row: utm.to_latlon(row['coordenada_x_local'],row['coordenada_y_local'],30,"T")[1] , 1) 
localesdf['radius']=5
```

```python
result = localesdf[['latitude',"longitude",'radius']].to_dict(orient="records")
```

```python
result[:5]
```

```python
jupyterviz.viz("tmp/map.htm","Idatamap",result[:100],900,900,{'center':[-3.7, 40],'scale':40000})
```
